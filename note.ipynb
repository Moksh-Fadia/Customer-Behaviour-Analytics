```python
import pandas as pd

df = pd.read_csv('data.csv')
```

```python
# category-wise median to preserve rating differences across product types and avoid global averaging
df['Review Rating'] = df.groupby('Category')['Review Rating'] .transform(lambda x: x.fillna(x.median()))

df.isnull().sum()
```

```python
# renaming columns (snake casing) for better documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})

df.columns
```

```python
# new column age group

bins = [18, 25, 40, 60, 100]
labels = [
    'Young (18–25)',
    'Adult (26–40)',
    'Middle_age (41–60)',
    'Senior (61–100)'
]

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

df[['age', 'age_group']].head(10)
```

```python
data_quality = pd.DataFrame({
    'missing_values': df.isna().sum(),
    'unique_values': df.nunique()
})

data_quality

# Note: Age group is a categorical feature thus missing values could not be safely replaced without knowing the correct category
```

```python
# mapping frequency to number of days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_freq_days'] = df['frequency_of_purchases'].map(frequency_mapping)

df[['purchase_freq_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)

df.columns
```

```python
df['customer_id'].duplicated().sum()     # checking for duplicate customer IDs
```

```python
age_spend_stats = (
    df.groupby('age_group', observed=True)['purchase_amount']
      .agg(mean_spend='mean', no_of_purchases='count', total_revenue='sum')
)

age_spend_stats
```

```python
# value vs volume

age_spend_stats.assign(
    revenue_share=lambda x: x['total_revenue'] / x['total_revenue'].sum()
)

# Insights:
# -Average spend per purchase is nearly the same across all age groups (~$60)
# -Revenue differences across age grps are driven by purchase frequency, not spending amount
# -Middle-aged customers (41–60) contribute the highest share of total revenue
# -Young users spend well per purchase but buy less frequently
# -Seniors provide stable but moderate revenue
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

age_spend_stats['mean_spend'].plot(kind='bar')

for i, count in enumerate(age_spend_stats['no_of_purchases']):
    plt.text(i, age_spend_stats['mean_spend'].iloc[i], f'n={count}', 
             ha='center', va='bottom', fontsize=9)

plt.title('Average Spend by Age Group')
plt.ylabel('Mean Purchase Amount (USD)')
plt.xlabel('Age Group')
plt.tight_layout()
plt.show()
```

```python
age_discount = (
    df.groupby(['age_group', 'discount_applied'])['purchase_amount']
      .mean()
      .unstack()
)

age_discount

# Insights:
# -Average spend is slightly higher without discounts across all age groups
# -Discounts do not increase per-purchase spending for any age group
```

```python
discount_rate_by_category = (
    df.groupby('category')['discount_applied']
      .value_counts(normalize=True)
      .unstack()
)

discount_rate_by_category
```

```python
# to check whether discounts increase or decrease spending; -ve lift means discounts reduce spending and +ve lift means discounts increase spending

discount_effect = (
    df.groupby(['category', 'discount_applied'])['purchase_amount']
      .mean()
      .unstack()
)

discount_effect['lift'] = discount_effect['Yes'] - discount_effect['No']

discount_effect

# Insights: In most categories, applying discounts leads to higher average purchase amounts
```

```python
discount_rate_by_category.plot(
    kind='bar',
    stacked=True,
    figsize=(6, 4)
)

plt.title('Discount Rate by Category')
plt.ylabel('Proportion')
plt.xlabel('Category')
plt.legend(title='Discount Applied')
plt.tight_layout()
plt.show()
```

```python
rating_stats_by_category = (
    df.groupby('category')['review_rating']
      .agg(avg_rating='mean', no_of_reviews='count')
      .sort_values(by='avg_rating', ascending=False)
)

rating_stats_by_category
```

```python
category_revenue = (
    df.groupby('category')['purchase_amount']
      .sum()
      .rename('total_revenue')
)

category_rating_revenue = (
    rating_stats_by_category
    .join(category_revenue)
)

category_rating_revenue

# Insights:
# -Ratings are almost the same across categories
# -Clothing earns the most revenue despite the lowest rating
# -Accessories have high review volume and strong revenue
# -Footwear is best rated but generates less revenue
# -Ratings do not directly correlate with revenue
```

```python
plt.figure(figsize=(6, 4))

rating_stats_by_category['avg_rating'].plot(kind='barh')

for i, count in enumerate(rating_stats_by_category['no_of_reviews']):
    plt.text(rating_stats_by_category['avg_rating'].iloc[i], i, f'n={count}', 
             va='center')

plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
df['is_repeat_customer'] = df['previous_purchases'] > 5

repeat_purchase_rate = df['is_repeat_customer'].mean()
repeat_purchase_rate

# Insights: Approx 89% of customers are repeat customers
```

```python
df.groupby('is_repeat_customer')['purchase_amount'].agg(
    mean_spend='mean',
    total_revenue='sum',
    no_of_purchases='count'
)

# Insights: Repeat customers contribute significantly more to total revenue compared to new customers
```

```python
df.groupby('is_repeat_customer')['review_rating'].agg(
    avg_rating='mean',
    count='count'
)
```

```python
df['customer_type'] = pd.cut(
    df['previous_purchases'],
    bins=[0, 1, 5, 10, 100],
    labels=['New', 'Occasional', 'Regular', 'Heavy']
)

df.groupby('customer_type')['purchase_amount'].agg(
    mean_spend='mean',
    total_revenue='sum',
    count='count'
)

# Insights:
# -Heavy customers generate most of the revenue due to very high purchase volume
# -Revenue growth is driven by frequency, not higher spend
# -New customers contribute little revenue in the short term
# -Business is highly dependent on loyal (heavy) customers
```

```python
df.groupby('is_repeat_customer')['purchase_amount'].sum().plot(
    kind='bar',
    title='Revenue Contribution: Repeat vs Non-Repeat Customers'
)
```

```python
# Estimating annual purchase amount per customer ie. how much a customer is likely to spend in a year based on their purchase frequency and amount

df['annual_purchase_estimate'] = (
    365 / df['purchase_freq_days']
) * df['purchase_amount']


df.groupby('customer_type')['annual_purchase_estimate'].agg(
    mean='mean',
    median='median',
    count='count'
)

# Insights:
# -Annual customer value is similar across all customer types
# -New customers’ high annual mean is driven by few high-spend outliers
# -Heavy customers show the most stable and reliable value
# -Overall, customer value varies more by individual behavior than by loyalty category
```

```python
# spend by subscription status

df.groupby('subscription_status')['purchase_amount'].agg(
    mean_spend='mean',
    purchase_count='count',
    total_revenue='sum'
)

# Insights:
# - Non-subscribers generate most of the revenue due to higher purchase volume
# - Subscription status does not increase per-transaction spending
# - Revenue differences are driven by purchase count, not spend size
```

```python
df.groupby('subscription_status')['is_repeat_customer'].mean()

# Insights:
# - Subscribers have a slightly higher repeat purchase rate than non-subscribers
# - Subscription status is positively associated with customer retention
# - The retention gap is minimal, suggesting subscriptions improve loyalty but not significantly
```

```python
# subscription status vs annual purchase estimate

df.groupby('subscription_status')['annual_purchase_estimate'].agg(
    mean='mean',
    median='median',
    count='count'
)


# Insights:
# - Subscribers have a slightly higher estimated annual value than non-subscribers
# - Median annual value is marginally higher for subscribers, indicating more consistent spending
# - Differences in annual value by subscription status are present but minimal
```

```python
# subscription status vs category spend

df.groupby(['category', 'subscription_status'])['purchase_amount'].mean().unstack()

# Insights:
# - Subscription impact on spend varies by category rather than being uniform
# - Subscribers spend slightly more in Footwear and Outerwear categories
# - Non-subscribers spend marginally more in Accessories and Clothing
```

```python
df.groupby('subscription_status')['annual_purchase_estimate'].sum() / df['annual_purchase_estimate'].sum()

# Insights:
# - Non-subscribers account for the majority of estimated annual purchase value
# - Subscribed customers contribute a smaller share of projected annual revenue
# - Revenue concentration currently favors non-subscribers despite higher subscriber retention
```

```python
# percentage revenue share by each category  

category_revenue_share = (
    df.groupby('category')['purchase_amount']
      .sum()
      .pipe(lambda x: (x / x.sum()) * 100)
      .rename('revenue_share')
)

category_revenue_share
```

```python
# percentage revenue share by age grp  

df.groupby('customer_type')['purchase_amount'] \
  .sum() \
  .pipe(lambda x: (x / x.sum()) * 100)
```
